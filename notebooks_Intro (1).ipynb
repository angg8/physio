```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = ['No submersion','Room Temp','Cold Water']
dom = [9,-1.4,12.7]
```

```python
X_axis = np.arange(len(X))
  
plt.bar(X_axis, dom, label = 'Condition of Apnea')
  
plt.xticks(X_axis, X)
plt.xlabel("Condition")
plt.ylabel("Percent change from restng HR")
plt.title("Percent change vs Conditon of Apnea")
plt.legend()
plt.show()
```
